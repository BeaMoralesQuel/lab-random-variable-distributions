```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numerical=pd.read_csv('numerical.csv')
numerical
```

## 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
for col in numerical.columns:
    sns.displot(numerical[col],kde=True)
    plt.show()
```

## 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.


```python
#Log Transformation
```

```python
num_transform=np.where(np.isfinite(np.log10(numerical)),np.log10(numerical),0)
```

```python
num_transform=pd.DataFrame(num_transform,columns=numerical.columns)
num_transform
```

```python
for col in num_transform.columns:
    sns.displot(num_transformed[col],kde=True)
    plt.show()
```
